```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import warnings
import seaborn as sns
from pandas.tseries.holiday import USFederalHolidayCalendar as calendar
```

```python
test_19 = pd.read_csv('../preprocessed_data/yellow_tripdata_2019-01.csv')
test_19 = test_19.sample(frac =.1).reset_index()
test_19
```

```python
test_19['time_hr'] = test_19['tpep_pickup_datetime'].str[11:13]
test_19['tpep_pickup_datetime'] = pd.to_datetime(test_19['tpep_pickup_datetime'])
test_19['date'] = test_19['tpep_pickup_datetime'].dt.normalize()
test_19 = test_19.drop(['VendorID', 'RatecodeID', 'store_and_fwd_flag', 'extra', 'mta_tax', 'improvement_surcharge', 'congestion_surcharge', 'index'],axis=1)
test_19
```

```python
test_weather = pd.read_csv('../preprocessed_data/test_weather.csv')
test_weather = test_weather.drop(['STATION'],axis=1)
test_weather['PRCP']= (test_weather['PRCP'] > 0)
test_weather['SNOW']= (test_weather['SNOW'] > 0)
test_weather['SNWD']= (test_weather['SNWD'] > 0)
test_weather['TAVG'] = 0.5*(test_weather['TMAX'] + test_weather['TMIN'])
test_weather = test_weather.drop(['TMAX','TMIN'],axis=1)
test_weather
```

```python
test_weather['date']  = pd.to_datetime(test_weather['DATE'])
test_weather = test_weather.drop(['DATE'],axis=1)
test_weather
```

```python
test_19 = test_19[test_19['payment_type'] == 1]
cal = calendar()
holidays = cal.holidays(start="2019-01-01", end="2019-01-31")
test_19['holiday'] = test_19['date'].isin(holidays)
test_19['date'] = pd.to_datetime(test_19['date'])
test_19['weekday'] = ((pd.DatetimeIndex(test_19.date).dayofweek) // 5 == 0).astype(float)
```

```python
test_19 = test_19.merge(test_weather, on='date')
test_19
```

```python
test_19=test_19.drop(['tpep_pickup_datetime', 'tpep_dropoff_datetime'],axis=1)
test_19['weekday'] = test_19['weekday'] == 1
test_19['Extreme'] = False
test_19
```

```python
test_19.to_csv('../preprocessed_data/test_19.csv')
```
