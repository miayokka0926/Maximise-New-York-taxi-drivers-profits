## Demand count by date

```python
y_s1['date'] = y_s1['tpep_pickup_datetime'].str[8:10]
count_date = y_s1['date'].value_counts().sort_index()

count_date
```

```python
count_date.index
```

```python
count_date = pd.DataFrame(count_date, index = count_date.index)
count_date.date = pd.to_numeric(count_date.date, errors='coerce')
count_date['count'] = count_date['date']
count_date = count_date.drop('date',axis=1)
count_date
```

```python
plt.bar(count_date.index, count_date['count'], color ='purple', width = 0.4)
 
plt.xlabel("date")
plt.ylabel("demand count")
plt.title("demand per day season-1 2018")
plt.savefig("../plots/demand per day season-1 2018.png")
plt.show()
```

## Demand count by region

# Sample

```python
sample = pd.read_feather('../preprocessed_data/sample.feather')
# this sample is generated in the "full model" notebook, and I read it here directly.
```

```python
DO_s1 = sample['PULocationID'].value_counts().sort_index()
DO_s1 = pd.DataFrame(DO_s1, index = DO_s1.index)
DO_s1
```

```python
gdf_s1 = gpd.GeoDataFrame(pd.merge(DO_s1, sf, left_on=DO_s1.index, right_on='LocationID')).drop('LocationID',axis=1)
gdf_s1['DO_count'] = gdf_s1['PULocationID']
gdf_s1.drop('PULocationID', axis=1, inplace = True)

gdf_s1
```

```python
y_var = "DO_count"
vmin, vmax = 0, gdf_s1["DO_count"].max()
fig, ax = plt.subplots(1, figsize=(11,11))

#title and annotation
ax.set_title('demand in each region')

#colorbar legend
sn = plt.cm.ScalarMappable(cmap = 'Blues', norm = plt.Normalize(vmin=vmin, vmax = vmax))

sn.set_array([])
fig.colorbar(sn)

gdf_s1.plot(column = y_var, cmap="Blues", linewidth=1.5,ax=ax,edgecolor = '0.4')
plt.savefig("../plots/demand in each region.png")
plt.show
```
